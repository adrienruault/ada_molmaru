For each  country, we have to get inside its directory, and get the data from each of the reports

- We move to "guinea_data" repositery  
- We get all the .csv files' title  and stock them into 'filenames_guinea'  
- We use a 'for' loop to get through all the files

```python
os.chdir("guinea_data" )
filenames_guinea = pd.Series(os.listdir(os.getcwd()))

#Initialization of the DataFrame 'guinea_data' for all 12 months of 2014 with 'Daily Average of Death' and 'Daily Average of New Cases' as Series
month_index = range(1,13)
zero_data = np.zeros(shape=(12,3))
guinea_data = pd.DataFrame(zero_data, columns=['NC','Death','Nb_report'])
guinea_data.index = month_index

#For each report, we add the number of deaths and new cases to the 'guinea_data_frame'
for title in filenames_guinea:
    
    #Read and replace NaN values in the reports by '0'
    mb = pd.read_csv(title).fillna(0)
    
    #Selection of the columns in the report
    mb = mb[['Date','Description','Totals']]

    #Selection of the rows in the report
    mb = mb[3:14] 
    mb = mb.drop(list(range(4,13)))
    
    #adding data from the reports to 'guinea_data'
    guinea_data.NC[float(title[5:7])] = guinea_data.NC[float(title[5:7])] + float(mb.iloc[0,2])
    guinea_data.Death[float(title[5:7])] = guinea_data.Death[float(title[5:7])] + float(mb.iloc[1,2]) 
    guinea_data.Nb_report[float(title[5:7])] += 1

#Average the totals by the number of reports    
for month in month_index:
    if guinea_data.Nb_report[month] != 0:
        guinea_data.NC[month] = guinea_data.NC[month] /guinea_data.Nb_report[month]
        guinea_data.Death[month] = guinea_data.Death[month] /guinea_data.Nb_report[month]



#Exit the repository
os.chdir("../" )

#Display
guinea_data
```

Now we do the same for the liberia folder

```python
os.chdir("liberia_data" )

#Initialization of the DataFrame 'liberia_data' for all 12 months of 2014 with 'Daily Average of Death' and 'Daily Average of New Cases' as Series
filenames_liberia = pd.Series(os.listdir(os.getcwd()))

#Initialization of the DataFrame 'liberia_data' for all 12 months of 2014 with 'Daily Average of Death' and 'Daily Average of New Cases' as Series
month_index = range(1,13)
zero_data = np.zeros(shape=(12,3))
liberia_data = pd.DataFrame(zero_data, columns=['NC','Death','Nb_report'])
liberia_data.index = month_index

#For each report, we add the number of deaths and new cases to the 'liberia_data_frame'
for title in filenames_liberia:
    
    #Read and replace NaN values in the reports by '0'
    mb = pd.read_csv(title).fillna(0)
    
    #Selection of the columns in the report
    mb = mb[['Date','Variable','National']]
    
    #Selection of the rows in the report
    mb = mb[7:26]
    mb = mb.drop(list(range(8,23)),axis = 0)
    
    #adding data from the reports to 'guinea_data'
    liberia_data.NC[float(title[5:7])] = liberia_data.NC[float(title[5:7])] + mb.iloc[1:4,2].sum()
    liberia_data.Death[float(title[5:7])] = liberia_data.Death[float(title[5:7])] + mb.iloc[0,2].sum()
    liberia_data.Nb_report[float(title[5:7])] += 1

#Average the totals by the number of reports    
for month in month_index:
    if liberia_data.Nb_report[month] != 0:
        liberia_data.NC[month] = liberia_data.NC[month] /liberia_data.Nb_report[month]
        liberia_data.Death[month] = liberia_data.Death[month] /liberia_data.Nb_report[month]



#Exit repository
os.chdir("../" )


#Display
liberia_data
```

```python
os.chdir("sl_data" )

#Initialization of the DataFrame 'sl_data' for all 12 months of 2014 with 'Daily Average of Death' and 'Daily Average of New Cases' as Series
filenames_sl = pd.Series(os.listdir(os.getcwd()))

#Creation du DataFrame pour la guinée
month_index = range(1,13)
zero_data = np.zeros(shape=(12,3))
sl_data = pd.DataFrame(zero_data, columns=['NC','Death','Nb_report'])
sl_data.index = month_index

#Pour chaque rapport, on ajoute les valeurs du nombre de mort, etc...
for title in filenames_sl:
    mb = pd.read_csv(title).fillna(0)
    #Section des colonne qui nous intéresse
    mb = mb[['date','variable','National']]
    #selection des lignes qui nous intéresse
    mb = mb[2:12]
    mb = mb.drop([5,6,7,8],axis = 0)
    #convert dtype from 'object' to 'float'
    mb[['National']]=mb[['National']].apply(pd.to_numeric)
    #opération de moyennage
    sl_data.NC[float(title[5:7])] = sl_data.NC[float(title[5:7])] + mb.iloc[0:3,2].sum()
    sl_data.Death[float(title[5:7])] = sl_data.Death[float(title[5:7])] + mb.iloc[4:6,2].sum()
    sl_data.Nb_report[float(title[5:7])] += 1

#On divise par le nombre de rapport pour chaque mois   
for month in month_index:
    if sl_data.Nb_report[month] != 0:
        sl_data.NC[month] = sl_data.NC[month] /sl_data.Nb_report[month]
        sl_data.Death[month] = sl_data.Death[month] /sl_data.Nb_report[month]



#On sort du répertoire de la guinée
os.chdir("../" )

#Display
sl_data[['NC','Death']]
```
