```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = 'https://www.topuniversities.com/sites/default/files/qs-rankings-data/357051.txt?_=1508694016501'
r = requests.get(URL)
```

```python
uni_dict = r.json()
```

create the data frame

```python
columns = ['Name','Region','Location','Rank','Score','Total staff', 'International staff', 'Total students', 'International students']
uni_rank = pd.DataFrame(index = [], columns=columns)
uni_rank
```

```python
def get_numbers_stud_staff(url):
    r_uni = requests.get("https://www.topuniversities.com" + url)
    
    html_request_uni = BeautifulSoup(r_uni.text, 'html.parser')
    
    header_staff = html_request_uni.find('div',class_='faculty-main wrapper col-md-4')
    header_stud_total = html_request_uni.find('div',class_='students-main wrapper col-md-4')
    header_stud_inter = html_request_uni.find('div',class_='int-students-main wrapper col-md-4')
    
    if (header_staff == None):
        return (0,0,0,0)

    subdiv_staff_total = header_staff.find('div', class_='total faculty')
    subdiv_staff_inter = header_staff.find('div', class_= 'inter faculty')
    subdiv_stud_total = header_stud_total.find('div', class_='total student')
    subdiv_stud_inter = header_stud_inter.find('div', class_='total inter')
    
    if (subdiv_staff_inter == None):
        return (0,0,0,0)
    
    num_staff_total = int(subdiv_staff_total.find('div', class_ = 'number').text[1:-1].replace(',',''))
    num_staff_inter = int(subdiv_staff_inter.find('div', class_ = 'number').text[1:-1].replace(',',''))
    num_stud_total = int(subdiv_stud_total.find('div', class_ = 'number').text[1:-1].replace(',',''))
    num_stud_inter = int(subdiv_stud_inter.find('div', class_ = 'number').text[1:-1].replace(',',''))
    
    return (num_staff_total, num_staff_inter, num_stud_total, num_stud_inter)
```

```python
uni_list = uni_dict['data']
i = 0
for uni in uni_list[:200]:
    i+=1
    print(i)
    (tot_staff, inter_staff, tot_stud, inter_staff) = get_numbers_stud_staff(uni['url'])
    uni_df = pd.DataFrame([[uni['title'],uni['region'], uni['country'], uni['rank_display'].replace('=',' '),
                            uni['score'], tot_staff, inter_staff, tot_stud, inter_staff]], columns = columns)
    uni_rank = uni_rank.append(uni_df)

uni_rank.index = list(range(1,201))
```

```python
URL = 'https://www.topuniversities.com/sites/default/files/qs-rankings-data/357051_indicators.txt'
r = requests.get(URL)
```

```python
uni_table = r.json()
```

```python
columns_table = uni_table['columns']
data_table = uni_table['data']
```

```python
features = []
features_title = []

for column in columns_table:
    soup = BeautifulSoup(column['title'], 'html.parser')
    features.append(soup.text)
    features_title.append(column['data'])
    
print(features)
print(features_title)
```

Create data frame for universities

```python
uni_rank_2 = pd.DataFrame(columns=features_title)

interesting_features = ['region', 'location', 'uni', 'rank', '2971070_rank_d', '2971074_rank_d']

for data in data_table[:200] :
    
    features_values = pd.DataFrame(columns= features_title)
    
    for feature in data :
        
        if data[feature]:
            soup = BeautifulSoup(data[feature], 'html.parser')
        else : soup = BeautifulSoup('Nan', 'html.parser')
        
        features_values.loc['0',feature] = soup.text
    
    uni_rank_2 = uni_rank_2.append(features_values)
    
uni_rank_2.columns = features
uni_rank_2.index = list(range(0,200))
uni_rank_2
```
